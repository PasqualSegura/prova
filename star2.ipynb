```python
import numpy as np
import matplotlib.pyplot as plt
from math import gcd
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def makeData (n, s, startingPoint):
    x = []
    y = []
    i = 0
    while i < n + 1:
        x.append( np.cos(startingPoint + np.pi/2 + 2*np.pi*s*i/n) )
        y.append( np.sin(startingPoint + np.pi/2 + 2*np.pi*s*i/n) )
        i+=1
    return [x,y]

def drawStar(n,s):
    numberOfSubStars = gcd(n,s)
    j = 0
    fig, ax = plt.subplots()
    ax.set_xlim((-1,1))
    ax.set_ylim((-1,1))
    x0,x1 = ax.get_xlim()
    y0,y1 = ax.get_ylim()
    ax.set_aspect(abs(x1-x0)/abs(y1-y0))
    ax.axis('off')
    while j < numberOfSubStars:
        tempData = makeData(n,s,2*np.pi*j/n)
        ax.plot(tempData[0],tempData[1])
        j+=1
    plt.show()   




```

```python
interactive_plot = interactive(drawStar,n=(5,18,1),s=(2,8,1))
output = interactive_plot.children[-1]
output.layout.height = '500px'
interactive_plot
```
